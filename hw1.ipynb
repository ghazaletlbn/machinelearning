```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/content/gdrive/My Drive/dataset/HW1_Dataset.csv')
```

```python
df['Density'] = df['2022']/df['Area (km²)']
df
#answer of q1
```

```python
aop = pd.DataFrame()
aop['Name'] = df['Name']
aop['avg of population'] = df.iloc[:,3:10].mean(axis=1)
print(aop)
#answer of q2
```

```python
df['2022 percent'] = (df['2022']/df.loc[:,'2022'].sum(axis=0))*100
df['2020 percent'] = (df['2020']/df.loc[:,'2020'].sum(axis=0))*100
df['2015 percent'] = (df['2015']/df.loc[:,'2015'].sum(axis=0))*100
df['2010 percent'] = (df['2010']/df.loc[:,'2010'].sum(axis=0))*100
df['2000 percent'] = (df['2000']/df.loc[:,'2000'].sum(axis=0))*100
df['1990 percent'] = (df['1990']/df.loc[:,'1990'].sum(axis=0))*100
df['1980 percent'] = (df['1980']/df.loc[:,'1980'].sum(axis=0))*100
df['1970 percent'] = (df['1970']/df.loc[:,'1970'].sum(axis=0))*100
df
#answer of q3
```

```python
gr = pd.DataFrame()
gr['Name'] = df['Name']
gr['growth rate'] = df['2022']/df['2020']
print(gr)
#answer of q4
```

```python
df =df.sort_values(by='2022 percent', ascending=False)
df
#answer of q5
```

```python
df['Rank'] = np.arange(df.shape[0])+1
df
#answer of q6
```

```python
df['binary population class'] = np.where(df['2022']>300000, 'more than 300', 'less than 300')
df
#answer of q7
```

```python
conditions = [
    (df['2022 percent'] < 2),
    (df['2022 percent'] >= 2) & (df['2022 percent'] < 5),
    (df['2022 percent'] >= 5)
    ]
values = [0,1,2]
df['population class'] = np.select(conditions, values)
df
#answer of q8
```

```python
df['growth status'] = np.where(df['2022']>df['1970'], 0, 1)
df
#answer of q9
```

```python
year = ['2022','2020','2015','2010','2000','1990','1980','1970']
plt.plot(year ,df.loc[[133],year].to_numpy().flatten())
plt.show()
#answer of q10
```
