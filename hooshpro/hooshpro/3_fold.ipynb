```python
import pandas as pd
df = pd.read_csv('train.csv')
import re
def clean_text(t):
    stop_words = {'با','که','از','در','می','تا','!','؟'}
    for word in stop_words:
         t=re.sub(str(word),"",str(t))
    return t
```

```python
def separate_by_class(df,arr):
    lable0 = dict()
    lable1 = dict()
    lable2 = dict()
    lable3 = dict()
    lable4 = dict()
    lable5 = dict()
    for i in range(len(arr)):
        vector = df.loc[arr[i]]
        words = clean_text(vector[1]).split()
        class_value = vector[-1]
        for n in range(len(words)):
            if words[n] not in lable0:
                lable0[words[n]]=1
            else:
                lable0[words[n]]= lable0[words[n]]+1
        if (class_value == 1):
            for j in range(len(words)):
                if words[j] not in lable1:
                    lable1[words[j]]=1
                else :
                    lable1[words[j]] = lable1[words[j]]+1
        elif (class_value == 2):
            for j in range(len(words)):
                if words[j] not in lable2:
                    lable2[words[j]]=1
                else :
                    lable2[words[j]] = lable2[words[j]]+1
        elif (class_value == 3):
            for j in range(len(words)):
                if words[j] not in lable3:
                    lable3[words[j]]=1
                else :
                    lable3[words[j]] = lable3[words[j]]+1                    
        elif (class_value == 4):
            for j in range(len(words)):
                if words[j] not in lable4:
                    lable4[words[j]]=1
                else :
                    lable4[words[j]] = lable4[words[j]]+1                    
        elif (class_value == 5):
            for j in range(len(words)):
                if words[j] not in lable5:
                    lable5[words[j]]=1
                else :
                    lable5[words[j]] = lable5[words[j]]+1                        
    return lable0,lable1,lable2,lable3,lable4,lable5
```

```python
def labeling(test,arr,l0,l1,l2,l3,l4,l5):
    testans = dict()
    for i in range(len(arr)):
        vector = df.loc[arr[i]]
        words = clean_text(vector[1]).split()
        p1 = [0 for i in range(len(words))]
        p2 = [0 for i in range(len(words))]
        p3 = [0 for i in range(len(words))]
        p4 = [0 for i in range(len(words))]
        p5 = [0 for i in range(len(words))]
        sum1 = sum (l1.values ())
        sum2 = sum (l2.values ())
        sum3 = sum (l3.values ())
        sum4 = sum (l4.values ())
        sum5 = sum (l5.values ())
        v = len(l0)
        for m in range(len(words)):
            if(words[m] not in l1):
                count1 = 1
            else:
                count1 = l1[words[m]] +1
            p1[m] = count1/(sum1+v)
        
            if(words[m] not in l2):
                count2 = 1
            else:
                count2 = l2[words[m]] +1
            p2[m] = count2/(sum2+v)  
        
            if(words[m] not in l3):
                count3 = 1
            else:
                count3 = l3[words[m]] +1
            p3[m] = count3/(sum3+v)
        
            if(words[m] not in l4):
                count4 = 1
            else:
                count4 = l4[words[m]] +1
            p4[m] = count4/(sum4+v)  
        
            if(words[m] not in l5):
                count5 = 1
            else:
                count5 = l5[words[m]] +1
                p5[m] = count5/(sum5+v)  
        
        Possibility1 = 1
        Possibility2 = 1
        Possibility3 = 1
        Possibility4 = 1
        Possibility5 = 1
        for i1 in p1 :  
            Possibility1 =  Possibility1*i1

        max = Possibility1 
    
        for i2 in p2 :
            Possibility2 =  Possibility2*i2
        if Possibility2 > max:
            max = Possibility2  
    
        for i3 in p3 :
            Possibility3 =  Possibility3*i3
        if Possibility3 > max:
            max = Possibility3  
     
        for i4 in p4 :
            Possibility4 =  Possibility4*i4
        if Possibility4 > max:
            max = Possibility4    
    
        for i5 in p5 :
            Possibility5 =  Possibility5*i5       
    
        if Possibility5 > max:
            testans[arr[i]] = 5
        elif max == Possibility4:
            testans[arr[i]]  = 4
        elif max == Possibility3:
            testans[arr[i]]  = 3
        elif max == Possibility2:
            testans[arr[i]] = 2
        else:
            testans[arr[i]]  = 1
        
    return testans
   
```

```python
#3fold
from numpy import array
from sklearn.model_selection import KFold
a =[0 for i in range(len(df))]
for i in range(len(df)):
    a[i] = i

kfold = KFold(3, True, 1)
for train, test in kfold.split(a):
    l0,l1,l2,l3,l4,l5 = separate_by_class(df,train)
    testanswers = labeling(df,test,l0,l1,l2,l3,l4,l5)
    count =0
    for z in test:
        vect= df.loc[z]
        if(vect[2] == testanswers[z]):
            count = count+1
        
    print(count)
print("size of test :",len(test))
    
```
